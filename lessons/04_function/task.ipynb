Исходные данные имеют следующую структуру:

1. перечень всех документов
```

    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},

```

```
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Общие требования к программе:
- код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
- в коде отсутствуют глобальные переменные (за исключением **documents** и **directories**);
- пользовательский ввод обрабатывается в цикле **while** до тех пор, пока пользователь явно не завершит программу (вводом команды "**q**").

````python
# list of all documents
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

# a list of shelves where documents are stored (if the document is in documents, then it must be in directories as well)
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

def main():
    while True:
        command = input('Введите команду:\n')

        if command == 'q':
            return
        elif command == 'p':
            print_p()
        elif command == 's':
            print_s()
        elif command == 'l':
            print_l()
        elif command == 'ads':
            perform_ads()
        elif command == 'ds':
            perform_ds()
        elif command == 'ad':
            perform_ad()
        elif command == 'd':
            perform_d()
        elif command == 'm':
            perform_m()
        else:
            print('Не известная команда')

        print('')

def print_p():
    number_document = input('Введите номер документа:\n')

    if not isDocumentInDirectory(number_document):
        print('Документ не найден в базе')
        return

    number_key, name_key = ['number', 'name']

    for document in documents:
        if number_key not in document \
                or name_key not in document \
                or document[number_key] != number_document:
            continue

        print(f'Владелец документа: {document[name_key]}')
        break

def print_s():
    number_document = input('Введите номер документа:\n')

    if not isDocumentInDirectory(number_document):
        print('Документ не найден в базе')
        return

    for directory_key in directories:
        if number_document in directories[directory_key]:
            print(f'Документ хранится на полке: {directory_key}')
            break

def print_l():
    print_all_documents('')

def perform_ads():
    new_shelf = input('Введите номер полки:\n')

    if new_shelf in directories.keys():
        shelves = ', '.join(directories.keys())
        print(f'Такая полка уже существует. Текущий перечень полок: {shelves}')
        return
    directories[new_shelf] = []
    shelves = ', '.join(directories.keys())
    print(f'Полка добавлена. Текущий перечень полок: {shelves}.')

def perform_ds():
    remove_shelf = input('Введите номер полки:\n')

    for directory_key in directories:
        if not remove_shelf in directory_key:
            continue

        if len(directories[directory_key]) > 0:
            shelves = ', '.join(directories.keys())
            print(f'На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: {shelves}.')
            return
        else:
            directories.pop(directory_key)
            shelves = ', '.join(directories.keys())
            print(f'Полка удалена. Текущий перечень полок: {shelves}.')
            return

    shelves = ', '.join(directories.keys())
    print(f'Такой полки не существует. Текущий перечень полок: {shelves}.')

def perform_ad():
    number_document = input('Введите номер документа:\n')
    type_document = input('Введите тип документа:\n')
    name = input('Введите владельца документа:\n')
    shelf = input('Введите полку для хранения:\n')

    if shelf not in directories:
        print('Такой полки не существует. Добавьте полку командой as')
        print_all_documents('Текущий список документов:')
        return

    for directory in directories.values():
        if number_document in directory:
            numbers = []
            for directory in directories.values():
                numbers += directory
            print(numbers)
            str_numbers = ', '.join(numbers)
            print(f'Такой номер на полках уже есть. Текущий перечень номеров: {str_numbers}.')
            return

    documents.append({
        'type': type_document,
        'number': number_document,
        'name': name
    })

    directories[shelf].append(number_document)

    print_all_documents('Документ добавлен. Текущий список документов:')

def perform_d():
    number_document = input('Введите номер документа:\n')

    if not isDocumentInDirectory(number_document):
        print('Документ не найден в базе.')
        print_all_documents('Текущий список документов:')
        return

    for directory in directories.values():
        if number_document in directory:
            directory.remove(number_document)
            break

    for index in range(len(documents)):
        if number_document in documents[index].values():
            documents.pop(index)
            break
    print('Документ удален.')
    print_all_documents('Текущий список документов:')

def perform_m():
    number_document = input('Введите номер документа:\n')
    shelf = input('Введите номер полки:\n')

    if shelf not in directories:
        shelves = ', '.join(directories.keys())
        print(f'Такой полки не существует. Текущий перечень полок: {shelves}')
        return

    if not isDocumentInDirectory(number_document):
        print('Документ не найден в базе.')
        print_all_documents('Текущий список документов:')
        return

    for directory_key in directories:
        if number_document in directories[directory_key]:
            directories[directory_key].remove(number_document)
            directories[shelf].append(number_document)
            break

    print('Документ перемещен.')
    print_all_documents('Текущий список документов:')



def isDocumentInDirectory(number_document: str) -> bool:
    for directory in directories.values():
        if number_document in directory:
            return True
    return False

def print_all_documents(prefix: str):
    type_key, number_key, name_key = ['type', 'number', 'name']

    if prefix != '':
        print(prefix)

    for document in documents:
        if type_key not in document \
                or number_key not in document \
                or name_key not in document:
            continue

        shelf = '0'
        for directory_key in directories:
            if document[number_key] in directories[directory_key]:
                shelf = directory_key

        print(f'№: {document[number_key]}, тип: {document[type_key]}, владелец: {document[name_key]}, полка хранения: {shelf}')

main()
````
