### Задание 2
Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

```python
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

Результат классификации запишите в отдельный столбец region.

````python
import pandas

def get_keywords() -> pandas.DataFrame:
    return pandas.read_csv('keywords.csv')

def get_geo_data() -> dict:
    return {
        'Центр': ['москва', 'тула', 'ярославль'],
        'Северо-Запад': ['петербург', 'псков', 'мурманск'],
        'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
    }

def transform_to_region(keyword: str, geo_data: dict) -> str:
    '''
    search for mention in 'keyword' cities from dictionary 'geodata'
    '''
    lower_word = keyword.lower()
    for key, cities in geo_data.items():
        for city in cities:
            if city in lower_word:
                return key
    return 'undefined'

def add_region_column(keywords: pandas.DataFrame, geo_data: dict) -> pandas.DataFrame:
    '''
    adds a region to the 'keywords' table
    '''
    keywords['region'] = keywords['keyword'].apply(transform_to_region, geo_data=geo_data)
    return keywords

keywords = add_region_column(
    get_keywords(), 
    get_geo_data()
)

# keywords.head(10)
keywords[keywords['region'] != 'undefined'].head(10)
````
