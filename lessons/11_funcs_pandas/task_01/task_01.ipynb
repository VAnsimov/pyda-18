### Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas

def get_movies() -> pandas.DataFrame:
    return pandas.read_csv('movies.csv')

def get_rating() -> pandas.DataFrame:
    rating = pandas.read_csv('ratings.csv')
    return rating[['rating', 'movieId']]

def transform_to_rating(movieId: int, rating: pandas.DataFrame) -> str:
    '''
    searches the 'rating' table for a movie with id 'movieId' and translates the rating into text format
    '''
    filter_rating = rating[rating['movieId'] == movieId]
    mean_rating = filter_rating.mean()['rating']
    
    if mean_rating <= 2:
        return 'низкий рейтинг'
    elif mean_rating <= 4:
        return 'средний рейтинг'
    else:
        return 'высокий рейтинг'

def add_rating_column(movies: pandas.DataFrame, rating: pandas.DataFrame) -> pandas.DataFrame:
    '''
    adds a rating to the 'movies' table
    '''
    movies['rating'] = movies['movieId'].apply(transform_to_rating, rating=rating)
    return movies
    
movies = add_rating_column(
    get_movies(), 
    get_rating()
)
movies.head(10)
```
